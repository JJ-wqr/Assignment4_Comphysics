```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x/2)

a = 0
b = 20
n = 1000

x = np.linspace(a,b,n+1)
y = f(x)
h = (b-a)/n

M = (h/2)*(y[0] + 2*np.sum(y[1:n]) + y[n])

plt.plot(x,y)
plt.show()

print("Mass =", M)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(r):
    return np.sqrt(3*r)*r

a = 0
b = 5
n = 1000

r = np.linspace(a,b,n+1)
y = f(r)
h = (b-a)/n

radial = (h/2)*(y[0] + 2*np.sum(y[1:n]) + y[n])
M = 2*np.pi*radial

plt.plot(r,y)
plt.show()

print("Mass =", M)
```

```python
import numpy as np
import matplotlib.pyplot as plt

k = 1600

def f(x):
    return k*x

a = 0.05
b = 0.10
n = 1000

x = np.linspace(a,b,n+1)
y = f(x)
h = (b-a)/n

W = (h/2)*(y[0] + 2*np.sum(y[1:n]) + y[n])

plt.plot(x,y)
plt.show()

print("Work =", W)
```

```python
import numpy as np
import matplotlib.pyplot as plt

rho = 1000
g = 9.8

def f(y):
    return rho*g*np.pi*(y/3)**2*(12-y)

a = 4
b = 12
n = 1000

y = np.linspace(a,b,n+1)
val = f(y)
h = (b-a)/n

W = (h/2)*(val[0] + 2*np.sum(val[1:n]) + val[n])

plt.plot(y,val)
plt.show()

print("Work =", W, "J")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sqrt(1 + (x*np.sqrt(x**2+2))**2)

a = 0
b = 1
n = 1000

x = np.linspace(a,b,n+1)
y = f(x)
h = (b-a)/n

L = (h/2)*(y[0] + 2*np.sum(y[1:n]) + y[n])

plt.plot(x,y)
plt.show()

print("Length =", L)
```
